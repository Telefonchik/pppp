## Задание 6
Выполните решение СЛАУ с помощью np.linalg.solve(a, b) (согласно варианту). Выполните проверку решения.\
image.png

```python
a = np.array([[45, 3.5, 7.4],
              [3.1, 6, 2.3],
              [0.8, 7.4, 5]])
b = np.array([25, -15, 64])
# Решаем систему уравнений Ax = b методом численного решения (LU-разложение)
x = np.linalg.solve(a,b)
display(x)
# Проверка решения: подставляем x обратно в систему
# np.dot(a, x) вычисляет матричное произведение A·x
# np.allclose() сравнивает результат с вектором b с учётом погрешности вычислений
check = np.allclose(np.dot(a, x), b) # Проверка
check
```

## Задание 7
Выполните решение СЛАУ матричным методом. Выполните проверку решения.

```python
# Вычисляем обратную матрицу A⁻¹ и умножаем её на вектор правых частей b
x = np.dot(np.linalg.inv(a), b)
display(x)
# np.allclose() сравнивает результат с исходным вектором b, допуская машинную погрешность
check = np.allclose(np.dot(a, x), b) # Проверка
check
```

## Задание 10
Решите систему нелинейных уравнений с помощью функций root() и fsolve().\
image.png

```python
def equations(vars):
    x, y = vars # Распаковка входного вектора переменных
    f1 = np.sqrt(x) * np.cos(y) - 5*x*y - 3
    f2 = 5*y / (x**2 + y**2 + 1) + x - y - 4
    return [f1, f2]

# Начальное приближение
initial_guess = [1, 1]

# Решение через scipy.optimize.root методом Левенберга-Марквардта ('lm')
# Алгоритм минимизирует сумму квадратов невязок: 0.5 * Σ(f_i)².
sol_root = sp.optimize.root(equations, initial_guess, method='lm')

# Вывод результата с точностью до 6 знаков после запятой
print("Решение (root):")
print(f"x = {sol_root.x[0]:.6f}, y = {sol_root.x[1]:.6f}")

# Решение через scipy.optimize.fsolve
# Использует модифицированный метод Ньютона-Рафсона с численным якобианом.
sol_fsolve = sp.optimize.fsolve(equations, initial_guess)

print("\nРешение (fsolve):")
print(f"x = {sol_fsolve[0]:.6f}, y = {sol_fsolve[1]:.6f}")
```
