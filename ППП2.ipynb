## Задание 3 ##


Проведите вычисления по заданной формуле при указанных значениях параметров  (в командном окне или отдельной ячейке Jupyter Notebook):
1) параметры заданы в коде,
2) параметры вводятся с клавиатуры (используйте функцию input)

```python
import math
m = -20.1
alpha_deg = 20
alpha_rad = math.radians(alpha_deg)
res1 = (1/12) * (m**3 * math.cos(alpha_rad)) / (math.sin(alpha_rad) + math.cos(alpha_rad))**3
print("Результат (вариант а):", res1)
m = float(input("Введите m (для варианта б это 1.133538): "))
alpha = float(input("Введите alpha (для варианта б это 1.48): "))
res2 = (1/12) * (m**3 * math.cos(alpha)) / (math.sin(alpha) + math.cos(alpha))**3
print("Результат (ввод с клавиатуры):", res2)
```

## Задание 4 ##

 Создайте функцию: а) с помощью ключевого слова def, б) с помощью ключевого слова lambda (lambda-функция), - для расчёта функции вида y=f(x) в заданной точке. Постройте график этой функции в диапазоне от 0 до x0 с шагом 0.1 с помощью библиотеки matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return math.sinh(x) + math.sin(x) - 1
x = 3.1
y = lambda x: math.sinh(x) + math.sin(x) - 1
print(f"Значение функции (lambda): {y(x)}")
print(f"Значение функции (def): {f(x)}")
```

```python
# Построение графика в диапазоне от 0 до x (3.1) с шагом 0.1
# создаем список x значений.
x_range = list(np.arange(0, 3.1, 0.1))
# Добавляем конечную точку 3.1 вручную, чтобы график точно доходил до x0
x_range.append(3.1)
# Вычисляем y для каждого x
y_range = [y(xi) for xi in x_range]

plt.figure()
plt.plot(x_range, y_range)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sinh(x) + sin(x) - 1')
plt.grid(True)
plt.show()
```

## Задание 8 ##

Подсчёт количества элементов меньших и больших некоторого числа в матрице произвольного размера. Функция на вход получает матрицу, и индексы строки, столбца элемента, а возвращает список из двух элементов, где первый - количество элементов меньше указанного, а второй - больше.

```python
from random import randint
width = randint(2,10)
height = randint(2,10)
matr = [[randint(0,10) for _ in range(width)] for ___ in range(height)]
print(*matr, sep = "\n")
pos_y, pos_x = map(int, input("Введите координаты \n").split())
bigger = []
smaller = []
number = matr[pos_x][pos_y]
print("Выбранное вами число -", number)
for i in range(height):
    for j in range(width):
        if matr[i][j] > number:
            bigger +=[matr[i][j]]
        elif matr[i][j] < number:
            smaller +=[matr[i][j]]
print(*bigger)
print(*smaller)
```

7. `В Python есть простые типы (int, float, str, bool) и структуры данных: list (изменяемый список), tuple (неизменяемый кортеж), dict (словарь пар ключ-значение) и set (множество уникальных элементов).`
