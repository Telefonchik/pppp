## Задание 2

* В наборе данных имеется информация о Возрасте, Поле, и Происхождении каждой жертвы. Есть ли связь между этими признаками? Для этого рассчитайте коэффициент корреляции между парами этих переменных методами Numpy, Scipy или Pandas (на выбор).
* Дайте ответ на вопрос: верно ли, что женщины чаще оказываются жертвами по сравнению с мужчинами?

```python
df['Victim Sex'] = df['Victim Sex'].replace({"M": 1, "F": -1})
df = df[~df['Victim Sex'].isin(['H', 'X', np.nan])]
df["Victim Sex"].unique()
for v_d in df["Victim Descent"].unique():
    subset = df[df["Victim Descent"] == v_d]
    print(f"Люди происхожения {v_d}")
    display(subset[['Victim Sex', "Victim Age"]].corr())
```

В районе V есть значимая корреляция, там жертвами преступлений являются старые женщины и молодые мужчины, в остальных местах нет четкой зависимости

## Задание 3

```python
df['Victim Age'].hist()
```

чаще всего жертвами становятся люди от 20 до 40

```python
density = sp.stats.gaussian_kde(df['Victim Age'].dropna())

x_vals = np.linspace(0,100, 1000) # Specifying the limits of our data
density.covariance_factor = lambda : .5 #Smoothing parameter

density._compute_covariance()
plt.plot(x_vals,density(x_vals))
plt.show()
```

В целом на графике так и показано, только с более точными числами - вероятность более 1%(т.е. большинство) - люди в возрасте от 15 до 60 лет

```python
female_data = df[df['Victim Sex'] == -1]
male_data = df[df['Victim Sex'] == 1]
```

```python
density = sp.stats.gaussian_kde(df[df['Victim Sex'] == -1]['Victim Age'].dropna())

x_vals = np.linspace(0,100, 1000) # Specifying the limits of our data
density.covariance_factor = lambda : .5 #Smoothing parameter

density._compute_covariance()
plt.plot(x_vals,density(x_vals))
plt.show()
```

Статистика для девушек отличается не сильно от общей, на пару лет раньше становятся жертвами, и более резкий рост до 30 и такое же резкое падение после 30

```python
density = sp.stats.gaussian_kde(df[df['Victim Sex'] == 1]['Victim Age'].dropna())

x_vals = np.linspace(0,100, 1000) # Specifying the limits of our data
density.covariance_factor = lambda : .5 #Smoothing parameter

density._compute_covariance()
plt.plot(x_vals,density(x_vals))
plt.show()
```
