## Задание 3. Создайте анкету с контактными данными клиента

```python
# Функция очистки всех полей ввода
def clear():
    ent_name.delete(0, tk.END)
    ent_surname.delete(0, tk.END)
    ent_address_1.delete(0, tk.END)
    ent_address_2.delete(0, tk.END)
    ent_city.delete(0, tk.END)
    ent_region.delete(0, tk.END)
    ent_post.delete(0, tk.END)
    ent_country.delete(0, tk.END)

# Функция сохранения данных в текстовый файл
def save():
   # Открываем файл в режиме перезаписи ("w"). Если файла нет, он создастся автоматически.
    f = open("client_data.txt", "w")
    # Считываем данные из полей методом .get(), добавляем перенос строки и записываем списком
    f.writelines([ent_name.get() + "\n",
                  ent_surname.get() + "\n",
                  ent_address_1.get() + "\n",
                  ent_address_2.get() + "\n",
                  ent_city.get() + "\n",
                  ent_region.get() + "\n",
                  ent_post.get() + "\n",
                  ent_country.get() + "\n"])
    f.close()

    # Дублируем код очистки полей после сохранения
    ent_name.delete(0, tk.END)
    ent_surname.delete(0, tk.END)
    ent_address_1.delete(0, tk.END)
    ent_address_2.delete(0, tk.END)
    ent_city.delete(0, tk.END)
    ent_region.delete(0, tk.END)
    ent_post.delete(0, tk.END)
    ent_country.delete(0, tk.END)

window = tk.Tk()
window.title("Задание 3. Создайте анкету с контактными данными клиента")
# Создаём два фрейма-контейнера для логического разделения интерфейса
frm_a = tk.Frame()# Будет содержать метки и поля ввода
frm_b = tk.Frame()# Будет содержать кнопки управления

# --- Создание меток (Labels) ---
# Параметр master=frm_a привязывает виджеты к первому фрейму
lbl_name = tk.Label(master = frm_a, text="Имя")
lbl_surname = tk.Label(master = frm_a,text="Фамилия")
lbl_address_1 = tk.Label(master = frm_a,text="Адрес 1")
lbl_address_2 = tk.Label(master = frm_a,text="Адрес 2")
lbl_city = tk.Label(master = frm_a,text="Город")
lbl_region = tk.Label(master = frm_a,text="Регион")
lbl_post = tk.Label(master = frm_a,text="Почтовый индекс")
lbl_country = tk.Label(master = frm_a,text="Страна")

# --- Создание текстовых полей (Entry) ---
# width=50 задаёт примерную ширину поля в символах моноширинного шрифта
ent_name = tk.Entry(master = frm_a, width = 50)
ent_surname = tk.Entry(master = frm_a, width = 50)
ent_address_1 = tk.Entry(master = frm_a, width = 50)
ent_address_2 = tk.Entry(master = frm_a, width = 50)
ent_city = tk.Entry(master = frm_a, width = 50)
ent_region = tk.Entry(master = frm_a, width = 50)
ent_post = tk.Entry(master = frm_a, width = 50)
ent_country = tk.Entry(master = frm_a, width = 50)

# --- Размещение меток с помощью менеджера grid ---
# grid использует табличную сетку: row (строка), column (колонка)
lbl_name.grid(row = 0, column = 0)
lbl_surname.grid(row = 1, column = 0)
lbl_address_1.grid(row = 2, column = 0)
lbl_address_2.grid(row = 3, column = 0)
lbl_city.grid(row = 4, column = 0)
lbl_region.grid(row = 5, column = 0)
lbl_post.grid(row = 6, column = 0)
lbl_country.grid(row = 7, column = 0)

# --- Размещение полей ввода в той же сетке ---
# Поля сдвинуты в колонку 1, чтобы выстроиться напротив меток
ent_name.grid(row = 0, column = 1)
ent_surname.grid(row = 1, column = 1)
ent_address_1.grid(row = 2, column = 1)
ent_address_2.grid(row = 3, column = 1)
ent_city.grid(row = 4, column = 1)
ent_region.grid(row = 5, column = 1)
ent_post.grid(row = 6, column = 1)
ent_country.grid(row = 7, column = 1)

# Встраиваем заполненный фрейм анкеты в главное окно (по умолчанию сверху)
frm_a.pack()

# --- Создание кнопок ---
# command=clear привязывает вызов функции при клике
btn_remove = tk.Button(master = frm_b, text = "Очистить", command = clear)
btn_save = tk.Button(master = frm_b, text = "Сохранить", command = save)

# pack(side=tk.LEFT) выстраивает кнопки горизонтально слева направо
btn_remove.pack(side = tk.LEFT)
btn_save.pack(side = tk.LEFT)

# Размещаем фрейм с кнопками под анкетой
frm_b.pack()
window.mainloop()
```

## Задание 5. Реализуйте интерфейс для одного из заданий

Окно должно содержать кнопку для открытия выбранного файла. \
Содержимое файла после открытия отображается в двух Label или Text (массивы x, y). \
Три окна Text позволяют задать начальное значение, шаг, конечное значение для расчёта промежуточных значений (необходима проверка корректности введённых данных). \
Результаты расчёта выводятся на графике в новом окне (plt.show): исходные точки отображаются чёрными квадратами, а новые – зелёными квадратами, соединёнными прямыми линиями.

```python
from tkinter import filedialog
import matplotlib.pyplot as plt
from functools import partial
import numpy as np
from scipy.interpolate import interp1d

def openfile():
    path = tk.filedialog.askopenfilename(filetypes=[("Текстовые файлы", "*.txt")])
    if path:
        print("Файл выбран:", path)
        interpol(path)

def interpol(path):
    with open(path, "r") as file:
        str_x = file.readline()
        str_y = file.readline()
        x = list(map(int, str_x.split()))
        y = list(map(int, str_y.split()))
    data_window = tk.Tk()
    data_window.title("Вывод данных с файла")

    tk.Label(master = data_window, text = "x =" + str_x).grid(row = 0, column = 0)
    tk.Label(master = data_window, text = "y =" + str_y).grid(row = 1, column = 0)

    tk.Label(master = data_window, text = "Начальное значение").grid (row = 0, column = 1)
    tk.Label(master = data_window, text = "Шаг").grid (row = 1, column = 1)
    tk.Label(master = data_window, text = "Конечное значение").grid (row = 2, column = 1)

    vcmd = data_window.register(is_valid)

    ent_begin = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))
    ent_diff = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))
    ent_end = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))

    btn_draw = tk.Button(master = data_window, text = "Нарисовать график", command = partial(draw_plot, ent_begin,  ent_diff, ent_end, x, y))

    ent_begin.grid (row = 0, column = 2)
    ent_diff.grid (row = 1, column = 2)
    ent_end.grid (row = 2, column = 2)
    btn_draw .grid (row = 3, column = 2)
    data_window.mainloop()

def is_valid(newval):
    if newval == "":
        return True
    try:
        float(newval)
        return True
    except ValueError:
        return False

def draw_plot(e_begin, e_diff, e_end, x, y):
    begin, diff, end = e_begin.get(),  e_diff.get(), e_end.get()
    if begin!= "" and diff!= "" and end!= "":
        print("Числа введены корректно")
        new_x = np.arange(float(begin), float(end), float(diff))
        plt.plot(x, y, 'ks') #Базовый график
        plt.plot(new_x, interp1d(x, y, kind='quadratic')(new_x), 'gs-') #Интерполированный квадратичным
        plt.show()
    else:
        print("Ошибка: введено не число!", begin, diff, end)
        return

file_window = tk.Tk()
file_window.title("Открытие файла")
btn_openfile = tk.Button(master = file_window, text = "Открыть файл", command = openfile)
btn_openfile.pack()
file_window.mainloop()

```
